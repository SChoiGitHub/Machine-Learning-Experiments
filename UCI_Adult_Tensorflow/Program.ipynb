```python
#Imports and data imports
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn import preprocessing

#Using UCI Machine Learning Repository's database.
#Using data from the "Adult Data Set" (https://archive.ics.uci.edu/ml/datasets/Adult)
#Dua, D. and Graff, C. (2019). UCI Machine Learning Repository [http://archive.ics.uci.edu/ml]. Irvine, CA: University of California, School of Information and Computer Science. 

#Reading CSV Data into Pandas
col_names = [
    "age",
    "workclass",
    "fnlwgt",
    "education",
    "education_num",
    "marital_status",
    "occupation",
    "relationship",
    "race",
    "sex",
    "capital_gain",
    "capital_loss",
    "hours_per_week",
    "native_country",
    "income"
]
raw_training_data = pd.read_csv(
    "./adult.data",
    names=col_names,
    na_values="?",
    sep=", ")
raw_testing_data = pd.read_csv(
    "./adult.test",
    names=col_names,
    na_values="?",
    sep=", ",
    skiprows=1)


setlist = []
def massage_data(df):
    #Drop unknown values
    df = df.dropna()
    discrete_columns = [
        "workclass",
        "education",
        "marital_status",
        "occupation",
        "relationship",
        "race",
        "sex",
        "native_country",
        "income"
    ]
    #Holand-Netherlands is not a One-Hot column in the 
    #testing features, so it will be removed from
    #the training set
    df = df[df.native_country != "Holand-Netherlands"]
    df['income'] = df['income'].map({
        '<=50K.' : '<=50K',
        '<=50K' : '<=50K',
        '>50K.' : '>50K',
        '>50K' : '>50K',
    })
    
    
    #One-hot encoding
    for col in discrete_columns:
        one_hot = pd.get_dummies(df[col])
        df = df.drop(col,axis=1)
        df = df.join(one_hot)
    
    setlist.append((set(df.columns.values)))
        
        
    #separating into features and labels
    features = df[df.columns[:-2]]
    labels = df[df.columns[-2:]]
    
    #Feature normalization
    min_max_scaler = preprocessing.MinMaxScaler()
    features = min_max_scaler.fit_transform(features)
    
    return features, labels

training_features, training_labels = massage_data(raw_training_data)
testing_features, testing_labels = massage_data(raw_testing_data)

print(training_features.shape)
print(testing_features.shape)
```

```python
def build_model():
    model = keras.Sequential([
        layers.Flatten(input_shape=(103,)),
        layers.Dropout(0.4),
        layers.Dense(103, activation='elu'),
        layers.Dropout(0.4),
        layers.Dense(2, activation='softmax')
    ])

    optimizer = keras.optimizers.Adam(learning_rate=0.01)
    loss = tf.keras.losses.CategoricalCrossentropy(from_logits=True)
    
    model.compile(loss=loss,
                optimizer=optimizer,
                metrics=['accuracy'])
    return model

model = build_model()
model.fit(training_features,training_labels,epochs=100)
testing_loss, testing_accuracy = model.evaluate(testing_features,testing_labels,verbose=2)
```
